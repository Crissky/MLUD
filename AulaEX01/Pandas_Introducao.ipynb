### Linhas

Filtrando linhas a partir de uma condição (linha que tem o valor 'grass' na coluna 'primary_type')

```
grass_df = df[df['primary_type'] == 'grass']

grass_df.head(5)
```

Filtrando por valor maior que

```
weak_against_water_df = df[df['against_water'] > 1]

weak_against_water_df[['english_name', 'primary_type', 'secondary_type', 'against_water']]
```

Filtrando por mais de uma coluna

```
weak_against_water_ground_df = df[(df['against_water'] > 1) & (df['against_ground'] > 1)]

weak_against_water_ground_df[['english_name', 'primary_type', 'secondary_type', 'against_water', 'against_ground']]
```

Filtrando por uma coluna ou outra

```
weak_against_water_or_ground_df = df[(df['against_water'] > 1) | (df['against_ground'] > 1)]

weak_against_water_or_ground_df[['english_name', 'primary_type', 'secondary_type', 'against_water', 'against_ground']]
```

Filtrando linhas onde a coluna 'mega_evolution' possui algum valor, ou seja, diferente de 'NaN'.

```
mega_df = df[df['mega_evolution'].notna()]

mega_df.head()
```
